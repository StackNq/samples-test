```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

n=int(input('Enter a number : '))
if n<0:
    print('Factorial of a negative number is not possible')
elif n==0:
    print('Factorial of 0 = 1')
else:
    a=factorial(n)
    print(f"Factorial of {n} = {a}")
```
